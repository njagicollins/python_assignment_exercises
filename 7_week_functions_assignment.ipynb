```python
#2. Write a function that takes first and last name and returns a dictionary.

def names_dict(first_name, last_name):
    """A simple function that takes names to return a dictionary
    its expected that both first and last names will be in string format
    """
    if not isinstance(first_name, str) or not isinstance(last_name, str):
        raise TypeError("Both inputs must be strings")
        
    full_name_dict = {
        'first_name' : first_name.title().strip(),
        'last_name' : last_name.title().strip(),      
    }

    return full_name_dict   

employee_names = [names_dict('collins', 'njagi'),
                  names_dict('brian', 'otieno'), 
                  names_dict('christine', 'mueni'),
                 ]
print(f"These are the full names of our employees: {employee_names}")
```

```python
#3. Write a function that accepts a list of numbers and returns their sum.
def summations(numbers):
    """A function that takes a list of numbers and returns their sum
    the inputs expected are float
    """
    added_numbers = sum(numbers)
    return added_numbers
    
total_profit = summations([300, 500, 200, 195.50, 400])
print(f"Our weekly total profit is: ${total_profit}")
```
