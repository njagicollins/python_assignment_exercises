```python
#Defination
#Dictionaries are a collection of keys-values pairs that can be changed and are not ordered
#Key values pairs where keys are the variables and the values as the data or information.
#keys are the box and values are the items in the box

#Simple example

collins_dictionary ={
    'first_name' : 'Collins',
    'last_name' : 'Njagi',
    'cell_phone': 9789834873,
    'country' : 'USA',
    'Profession' : 'Data Scientist'}

collins_dictionary

```

```python
#Acessing values in a dictionary
collins_dictionary['cell_phone']
print("\n")
collins_dictionary['Profession']
print("\n")
collins_dictionary['country']
```

```python
#Adding Key-value pairs in a dictionary
#I forgot to add learning institution to collins_dictionary. How can we remedy this?

collins_dictionary['University'] = 'Merrimack College'
collins_dictionary['County'] = 'Middle-sex'

collins_dictionary #this is calling the dictionary to confirm the updates
```

```python
#Adding key-values starting with an empty dictionary. Inserting key-value pairs from an empty dictionary

highland_apt_dictionatry = {}
highland_apt_dictionatry['tenant'] = 'Collins'
highland_apt_dictionatry['rent_amount'] = 1650
highland_apt_dictionatry['start_date'] = '17 JAN 25'
highland_apt_dictionatry['contract_duration'] = '13 months'

highland_apt_dictionatry
```

```python
#Modifying Values in a dictionary (how to edit or change values you already inserted in the dictionary)
#assuming the rent amount was 1550 instead ho can we correct this

highland_apt_dictionatry['rent_amount'] = 1550
highland_apt_dictionatry
```

```python
#Removing key-values pairs from the dictionary
#for example contract duraction is no longer an important information in our dictionary

#Method 1: Use of delete (del)

del highland_apt_dictionatry['contract_duration']
highland_apt_dictionatry
```

```python
#method two use of pop to remove rent amount 
highland_apt_dictionatry.pop('rent_amount')
highland_apt_dictionatry
```

```python
collins_dictionary.pop('County') #when using pop to remove key-pair, pop function returns the item you want
#to remove without necessarily calling the dictionary
```

```python
#Accessing Values in the dictionary in a profectional way
#for example if we want to access county information in collins dictionay

collins_dictionary.get('County', 'Not available') #get function reduces the self doubt
#for example if you forgot you had removed an item from the dictionary
#get function will help remember by adding 'Not available' otherwise
#you might think maybe you missed a comma or used the wrong brackets

collins_dictionary.get('first_name', 'Not available')
```

```python
#Looping through the dictionary to get all the keys

for Key_infor in sorted(collins_dictionary.keys(), reverse = True):
    print(f"\t - {Key_infor.title().strip()}")
```

```python
#looping through the dictionary to get all values
for values in collins_dictionary.values():
    print(F" \t - {values}")
```

```python
#looping through the dictionary to get all values
for values in collins_dictionary.values():
     if type(values) == int: #type is a function to refer to data types(int, float, string, boolean etc)
          print(values)
     else:
         print(f"{values.title().strip()}")
#conditional statement was added to include int in title format otherwise int does not
#take title() or strip() functions
```

```python
#Looping through the dictionary to get keys-values pairs

for keys, values in collins_dictionary.items():
    print(F"{keys.title().strip()} : {values}")
```
