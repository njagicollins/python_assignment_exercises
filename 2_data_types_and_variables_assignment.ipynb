```python
#### Assignment 3: Arithmetic Operations

#Write a Python program that:

#- Stores two numbers in variables.
#Assuming the value of x = 10 and the value of y = 5
#Calculates and prints:

#- Their sum
#- Their difference
#- Their product
#- Their division result
#Ensure your output is clear and easy to read.

x = 10
y = 5

print("Sum:", x + y)
print("Difference:", x - y)
print("Product:", x * y)
print("Division:", x / y)
```

```python
#### Assignment 4: String Manipulation
#Given the following variable:
#full_name = "  collins njagi  "
#Write code that:
#Removes the leading and trailing spaces.
#Converts the cleaned name to:
  #- Uppercase
  #- Lowercase
  #- Title case
#Prints each result on a new line.

full_name = " collins njagi "

print(f"My full name is: {full_name.strip().upper()}")
print(f"my full name is: {full_name.strip().lower()}")
print(f"my full name is: {full_name.strip().title()}")
```
