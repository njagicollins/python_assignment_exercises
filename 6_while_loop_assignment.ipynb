```python
#Practical Exercises

#1. Write a program that repeatedly asks the user for numbers and prints their sum. Stop when the user types 'quit'.

#prompt = input("Enter a number, type 'quit' to stop: ") when i use this prompt at this location, for some reason i have to 
#input 5 three times for it to to sum total 10!!

sum_of_numbers = 0

while True:
    user_input = input("Enter a number, type 'quit' to stop: ")
    if user_input.lower() == 'quit':
        break
    sum_of_numbers += float(user_input)


print(f"The sum of numbers you entered is: {sum_of_numbers}")
 
```

```python
#2. Create a while loop that removes numbers less than 10 from a list until all numbers are 10 or above.

age_group1 = [5, 11, 7, 14, 12, 11, 9, 8, 13]
age_ten_and_above = []

while age_group1:
    current_age = age_group1.pop(0)
    if current_age >= 10:
        age_ten_and_above.append(current_age)
    else:
        print(f"Kids below age 10 removed from the list: {current_age}")

print("final list (10 +): ", age_ten_and_above)
```
