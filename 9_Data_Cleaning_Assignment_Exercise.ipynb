3.1 Exercises: String Columns

1.Converting and Trimming Strings
    
    Assume a column Name has inconsistent data: " Alice ", "BOB", "Charlie " 
    Convert all names to string type.
    Remove extra spaces using .str.strip().
    Convert all names to title case.


```python
import pandas as pd

data_names = ['Alice', 'BOB', 'Charlie'] #creating names as a list 

names_df = pd.DataFrame(data = data_names, columns = ['Name']) #creating a DataFrame with name column
names_df.head()
```

```python
#cleaning the name column in the created df
names_df['Name_clean'] = names_df['Name'].str.strip().str.title()
names_df
```

```python
#to confirm no whitespaces and the code was successful i will use .apply(repr) it shows whitespaces, tab, indent
#For title case i can confirm with my visible but i can also use (.istitle()) just to confirm. for data type 
#i could use .dtype

names_df['Name_clean'].apply(repr) #this checks for whitespaces
```

```python
names_df['Name_clean'].dtype #confirms that our name column is in string format
```

2. Replacing Characters and Removing Special Characters

    Column Code contains values: "A-123", "B_456", "C@789!"
    Replace "-" with "_".
    Remove special characters like @ and ! using regex.

```python
codes =['A-123', 'B_456', 'C@789!']
             
names_df['Id_codes'] = codes

names_df.head()
```

```python
# Replace "-" with "_".
# method 1

names_df['Id_codes_cleaned'] = names_df['Id_codes'].str.replace('-', '_', regex = True).str.replace('@', '_', regex = True).str.replace('!', '', regex = True)

names_df[['Id_codes_cleaned', 'Id_codes']]
```

```python
#Remove special characters like @ and ! using regex.
#method 2
names_df['Id_codes'].str.strip().replace('-', '_', regex = True).str.replace(r'(\W)', '', regex = True) 

#'(\W)' replaces non-word characters, that is @ and !
```

3. Standardizing Categorical Values

    Column Gender has values: "M", "F", "male", "Female"
    Standardize all values to "male" and "female".

```python
#Column Gender has values: "M", "F", "male", "Female"

genders = ['M', 'F', 'male', 'female']

names_df = names_df.reindex(range(len(genders))) #reindex fixes names_df to accomodate the 
#fourth row brought about by genders
names_df['Gender'] = genders
names_df.head()
```

```python
#Standardize all values to "male" and "female"

names_df['Gender_clean'] = names_df['Gender'].str.lower().replace({'m':'male', 'f':'female'}) #I used a dictionary because i was replacing
#more than one value that is 'M' and 'F'

names_df[['Gender', 'Gender_clean']]
```

4. Extracting Text and Checking Length

    Column ProductCode has values: "AB1234", "CD5678", "EF9012"
    Extract the numeric part using .str.extract().
    Check the length of each string in ProductCode

```python
productcode = ["AB1234", "CD5678", "EF9012", None] # i will include None because 
#names_df was expanded by Gender column

names_df['Product_code'] = productcode
names_df.head()
```

```python
 #Extract the numeric part using .str.extract().

names_df['Product_code_clean'] = names_df['Product_code'].str.extract(r'(\d+)') 

# '\d' extracts digits 
#the '+' means we are extracting more than one digit

names_df[['Product_code', 'Product_code_clean']]
```

```python
#Check the length of each string in ProductCode
names_df['Product_code_clean'].str.len()
```

1.Converting to Numeric and Handling Errors

    Column Age contains: 25, "30", "unknown", 40
    Convert the column to numeric using pd.to_numeric() with errors='coerce'.
    Identify the resulting missing value.

```python
age = [25, "30", "unknown", 40]

names_df['Age'] = age
names_df.info() #by using .info() we can see that Age has been created as an object
#that is we don't know whether its str, float, int 
```

```python
#Convert the column to numeric using pd.to_numeric() with errors='coerce'.

names_df['Age_clean'] = pd.to_numeric(names_df['Age'], errors = 'coerce')

names_df.info()
```

```python
#Identify the resulting missing value.

names_df['Age_clean'].isna()
```

```python
names_df[names_df['Age_clean'].isna()] #deep diving to check which row has missing value
```

```python
names_df['Age_clean'].unique() #this is also can help us identify the missing value
```

2. Rounding and Replacing Values

    Column Price contains: 1.234, 2.567, 3.891
    Round all prices to 2 decimal places.
    Replace any negative values with 0

```python
price = [1.234, 2.567, 3.891, -2.375]

names_df['Price'] = price

names_df.head()
```

```python
#Round all prices to 2 decimal places.

names_df['Price_clean'] = names_df['Price'].round(2)

names_df[['Price', 'Price_clean']]
```

```python
#Replace any negative values with absolute

names_df['Price_clean'].abs() 

#in this case using absolute would be the best solution
#since there can never be a negative price unless it was meant to indicate the product was bought at a debt
#or was a loss or it could be an obvious typo!
```

```python
#Replace any negative values with 0
#method 1 using .clip(lower=0)

names_df['Price_clean'].clip(lower = 0)
```

```python
#Replace any negative values with 0

#method 2 using '.loc()'

names_df.loc[names_df['Price_clean'] < 0, 'Price_clean'] = 0

names_df[['Price', 'Price_clean']]
```
