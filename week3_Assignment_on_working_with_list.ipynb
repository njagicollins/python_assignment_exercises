```python
### Exercise 3: Looping Through a List

#Create a list called `students` containing at least five student names.

#Tasks:
#1. Use a `for` loop to print each name.
#2. Modify the loop to print a message such as:  
#  `"Welcome, <name>!"`

students = ['Collins', 'Jane', 'mueni', 'brian', 'SHARON']
for name in students:
    name = name.title().strip()
    print(F"Welcome,{name}!\n")
```

```python
### Exercise 4: Using `upper()` and `lower()` with Lists

#Create a list called `names` containing names written in lowercase.

#Tasks:
#1. Loop through the list and print each name in uppercase.
#2. Loop through the list again and print each name in title case.

names = ['collins', 'mbuemo', 'amad','obama','muthoni']
print(F"Legendary names in uppercase:")
for name in names:
    name = name.upper()
    print(name)
print("\n")
print(F"Names in Proper case:")
for name in names:
    name = name.title()
    print(name)
```

```python
### Exercise 5: List Length and Indexing

#Create a list called `scores` containing numeric values.

#Tasks:
#1. Print the number of elements in the list.
#2. Print the score at index 2.
#3. Check whether the list has more than five elements and print an appropriate message.

scores = [90, 85, 88, 92, 95, 96]
print(f"The following is the scores list:\n{scores}")
print("\n")
print(f"Index two in the scores list is:\n{scores[2]}")
print("\n")
if len(scores) >= 5:
    print(f"The list of scores has more than five elements")
```

```python
### Exercise 6: Slicing a List

#Create a list called `daily_sales` containing sales amounts for at least seven days.

#Tasks:
#1. Print the first three sales values.
#2. Print the last three sales values.
#3. Print all values except the first and last.

daily_sales = ['$200', '$450', '$350', '$500','$700', '$400', '$150']
print(f"The first three days of Orencia sales in the second Week of January 2026:\n{daily_sales[:3]}")
print("\n")
print(f"The last three days of Orencia sales in the second Week of January 2026:\n{daily_sales[-3:]}")
print("\n")
print(f"All Orencia sales in the 2nd week of january 2026, except the first and last sale:\n{daily_sales[1:6]}")
```

```python
### Exercise 7: Adding and Removing Elements

#Create a list called `tasks` representing daily tasks.

#Tasks:
#1. Add two new tasks to the list.
#2. Remove one task using its value.
#3. Remove the last task using a list method(index).
#4. Print the final list.
print(f"My daily tasks as data scientist:")
tasks = ['data cleaning', 'preprocessing', 'EDA', 'Model building']
print(tasks)
print("\n")
tasks.append('collaboration & communication')
tasks.append('research & updates')
print(f"My updated daily tasks as a data scientist:\n{tasks}")
print("\n")
tasks.remove('data cleaning')
print(f"My updated list of daily tasks after removing data cleaning:\n{tasks}")
print("\n")
tasks.pop(-1)
print(f"My final list after all the updates:\n{tasks}")
```

```python
### Exercise 8: Sorting and Ordering Lists

#Create a list called `product_prices` with numeric values.

#Tasks:
#1. Print the list in ascending order.
#2. Print the list in descending order.
#3. Print the original list to confirm it has not changed. 

product_prices = [25, 50, 17.25, 67, 9.95, 45]
ascending_product_prices_order = sorted(product_prices)
print(f"Product prices sorted in Ascending order:\n{ascending_product_prices_order}")
print("\n")
descending_product_prices_order = sorted(product_prices, reverse = True)
print(f"Product prices sorted in Descending order:\n{descending_product_prices_order}")
print("\n")
print(f"This is the original list of product prices:\n{product_prices}")
```
