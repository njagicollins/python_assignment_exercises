```python
### Excercise:

#1. Write a grading system:

#- 80 and above → A
#- 60–79 → B
#- 40–59 → C
#- Below 40 → Fail

grade = int(input(F"Please enter the grade: "))

if grade >= 80:
    print(f"Grade A")
elif grade >= 60 <= 79:
    print(f"Grade B")
elif grade >= 40 <= 59:
    print(f"Grade C")
else:
    print(F"Fail")
```

```python
#2. Write a program that prints the type of day:
#- Monday → "Start of the week"
#- Friday → "Almost weekend"
#- Saturday or Sunday → "Weekend"
#- Any other day → "Normal day"

day_of_the_week = str(input(f"Enter the day of the week: ")).lower().strip()

if day_of_the_week == 'monday':
    print("Start of the week")
elif day_of_the_week == 'friday':
    print("Almost weekend")
elif day_of_the_week in ['saturday', 'sunday']:
    print("Its weekend")
else:
    print("Normal day")
```
