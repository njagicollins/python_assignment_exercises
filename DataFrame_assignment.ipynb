1.  **Understanding DataFrames**\
    Create a table (DataFrame) to represent the following data:

| Student | Grade | Subject |
| --- | --- | --- |
| John | 88 | Math |
| Mary | 92 | English |
| Steve | 79 | Science |
| Linda | 85 | History |

-   Identify the **rows** and **columns**.

-   What will be the **default index**?

````python
#create a list of student scores
student_scores = [
    ['Jonh', 88, 'Math'],
    ['Mary', 92, 'English'],
    ['Steve', 79, 'Science'],
    ['Linda', 85, 'History']
]

#Call the list
student_scores


#import the necessary library
import pandas as pd


#convert the list into a DataFrame
df_student_scores = pd.DataFrame(data = student_scores, columns = ['Student', 'Grade', 'Subject'])


#call the DataFrame
df_student_scores.head()
````

````python
print('The rows and columns of the DataFrame are: ')
df_student_scores.shape
````

Default index will be 0 to 3

1.  **Custom Index**\
    Set the `Student` column as the **row index**.

    -   How does the DataFrame change?

    -   Print the DataFrame to see the new index.

````python
df_student_scores_indexed = df_student_scores.set_index('Student')
df_student_scores_indexed #the student column drops to a lower level as compared to the other columns
````

1.  **From Lists**\
    Create a DataFrame from this list of lists:
```python
data = [

    ['Banana', 80, 0.2],

    ['Cherry', 200, 1.5]

]
```
-   Set column names to `['Fruit', 'Quantity', 'Price']`.

1.  **Boolean Filtering**

-   Select all fruits with `Quantity > 100`.

-   Select all fruits with `Price > 0.5` **AND** `Quantity < 150`.

-   Select all fruits with `Price < 1` **OR** `Quantity > 150`.

````python
#Select all fruits with `Quantity > 100`

fruits_shopping_df[fruits_shopping_df['Quantity'] > 100]
````

````python
#Select all fruits with `Price > 0.5` **AND** `Quantity < 150`.

fruits_shopping_df[(fruits_shopping_df['Price'] > 0.5) & (fruits_shopping_df['Quantity'] < 150)]
````

````python
fruits_shopping_df
````

````python
#Select all fruits with `Price < 1` **OR** `Quantity > 150`.

fruits_shopping_df[(fruits_shopping_df['Price'] < 1) | (fruits_shopping_df['Quantity'] > 150)]
````

1.  **Copy vs View**

-   Create a subset of the fruit DataFrame with only `Fruit` and `Price` columns using `.copy()`.

-   Increase the `Price` by 10% in the subset.

-   Verify that the original DataFrame is **not affected**. 

````python
subset_df = fruits_shopping_df[['Fruit', 'Price']].copy()

subset_df 
````

````python
#increasing the price in the subset by 10% will be current price = 100%, new price after increase
#will be 100% + 10% , thus 110% which is 110% change into decimals = 1.1


subset_df['Price'] *= 1.1
subset_df
````

````python
#calling the original df to show that it wasn't affected by 10% price increase done on the copy
fruits_shopping_df
````
